## Theoretical Questions: -

Q1. What is unsupervised learning in the context of machine learning?

A1. Unsupervised learning is a type of machine learning where the system learns from data that does not have any labels or predefined outputs. It tries to find patterns, similarities, or structures within the data on its own without any guidance.

Q2. How does K-Means clustering algorithm work?

A2. K-Means clustering works by dividing data into a fixed number of groups called clusters. It starts by selecting some points as cluster centers, then assigns each data point to the nearest center. After that, it updates the centers by taking the average of all points in each cluster. This process repeats until the centers do not change much, forming stable clusters.

Q3.  Explain the concept of a dendrogram in hierarchical clustering?

A3. A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are grouped together. It displays the merging process of clusters step by step, where each branch represents a cluster and the height shows the distance or similarity between clusters.

Q4. What is the main difference between K-Means and Hierarchical Clustering?

A4. The main difference is that K-Means requires the number of clusters to be decided in advance, while Hierarchical Clustering does not. K-Means forms clusters by repeatedly updating centers, whereas Hierarchical Clustering builds a tree of clusters by either merging or splitting them step by step.

Q5. What are the advantages of DBSCAN over K-Means?

A5. DBSCAN can find clusters of any shape and can also identify noise or outliers, which K-Means cannot. It does not require specifying the number of clusters in advance and works well with unevenly sized or dense data.

Q6. When would you use Silhouette Score in clustering?

A6. Silhouette Score is used to measure how well the data points fit within their clusters. It helps evaluate the quality of clustering by checking how similar each point is to its own cluster compared to other clusters.

Q7. What are the limitations of Hierarchical Clustering?

A7. Hierarchical clustering is slow and less efficient for large datasets. It is sensitive to noise and outliers, and once clusters are merged or split, the process cannot be undone. It also does not always give the best clustering results for complex data.

Q8. Why is feature scaling important in clustering algorithms like K-Means?

A8. Feature scaling is important in K-Means because it ensures that all features contribute equally to the distance calculation. Without scaling, features with larger values can dominate the clustering process, leading to inaccurate or biased results.

Q9. How does DBSCAN identify noise points?

A9. DBSCAN identifies noise points as those that do not have enough neighboring points within a given distance. These points do not belong to any cluster because they fail to meet the minimum number of points required to form one.

Q10. Define inertia in the context of K-Means?

A10. Inertia in K-Means is the measure of how close the data points in a cluster are to their cluster center. It is calculated as the sum of squared distances between each point and its cluster center. Lower inertia means better and tighter clustering.

Q11. What is the elbow method in K-Means clustering?

A11. The elbow method is a technique used to find the optimal number of clusters in K-Means. It involves plotting the inertia against different numbers of clusters and looking for a point where the decrease in inertia slows down, forming an “elbow.” That point suggests the best number of clusters.

Q12. Describe the concept of "density" in DBSCAN.

A12. In DBSCAN, density refers to how closely data points are packed together in a region. A dense area has many points within a certain distance (called epsilon), and such regions are considered clusters, while sparse areas are treated as noise.

Q13. Can hierarchical clustering be used on categorical data?

A13. Yes, hierarchical clustering can be used on categorical data, but it requires a suitable distance measure like Hamming or Jaccard distance instead of Euclidean distance. This helps to correctly measure similarity between categorical values.

Q14. What does a negative Silhouette Score indicate?

A14. A negative Silhouette Score indicates that the data points may be assigned to the wrong clusters. It means the points are closer to points in other clusters than to those in their own cluster, showing poor clustering quality.

Q15. Explain the term "linkage criteria" in hierarchical clustering.

A15. Linkage criteria in hierarchical clustering decide how the distance between two clusters is calculated. It determines which clusters should be merged at each step based on methods like single linkage (minimum distance), complete linkage (maximum distance), or average linkage (average distance).

Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

A16. K-Means may perform poorly on data with varying cluster sizes or densities because it assumes all clusters are spherical and of similar size. As a result, it can incorrectly group points or split clusters unevenly when clusters have different shapes or densities.

Q17. What are the core parameters in DBSCAN, and how do they influence clustering?

A17. The core parameters in DBSCAN are eps (epsilon) and minPts (minimum points). Eps defines the maximum distance between two points to be considered neighbors, and minPts sets the minimum number of points needed to form a dense region. These parameters control how clusters are formed and how noise is detected.

Q18. How does K-Means++ improve upon standard K-Means initialization?

A18. K-Means++ improves standard K-Means by choosing initial cluster centers in a smarter way. It selects the first center randomly and then picks the next centers farther apart from the existing ones. This helps in faster convergence and reduces the chances of poor clustering results.

Q19. What is agglomerative clustering?

A19. Agglomerative clustering is a type of hierarchical clustering that builds clusters from the bottom up. It starts with each data point as its own cluster and then repeatedly merges the closest clusters until all points belong to a single cluster or the desired number of clusters is reached.

Q20. What makes Silhouette Score a better metric than just inertia for model evaluation?

A20. Silhouette Score is better than inertia because it considers both the compactness within clusters and the separation between different clusters. Inertia only measures how close points are to their cluster centers, while Silhouette Score gives a clearer idea of how well-defined and distinct the clusters are.

Q21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=1)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=1)
kmeans.fit(X)
labels = kmeans.labels_

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='red', marker='X', s=200)
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, random_state=1)

# Evaluate silhouette scores for k = 2 to 5
scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=1)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Display as a bar chart
plt.bar(range(2, 6), scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K = 2 to 5")
plt.show()
```

Q31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data with overlapping clusters
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X)
labels = kmeans.labels_

# Create decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot clusters and boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='red', marker='X', s=200)
plt.title("K-Means Clustering with Decision Boundaries")
plt.show()
```

Q44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=1)
labels = kmeans.fit_predict(X)

# Compute silhouette scores for each sample
silhouette_vals = silhouette_samples(X, labels)

# Plot silhouette coefficient for each sample
plt.figure(figsize=(8, 4))
y_lower = 10
for i in range(3):
    cluster_silhouette_vals = silhouette_vals[labels == i]
    cluster_silhouette_vals.sort()
    cluster_size = len(cluster_silhouette_vals)
    plt.barh(range(y_lower, y_lower + cluster_size),
             cluster_silhouette_vals, height=1)
    y_lower += cluster_size

plt.title("Silhouette Coefficient for each Sample (Iris Dataset)")
plt.xlabel("Silhouette Coefficient Value")
plt.ylabel("Samples")
plt.axvline(silhouette_score(X, labels), color="red", linestyle="--")
plt.show()
```

Q48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X_tsne)

# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10')
plt.title("Agglomerative Clustering on Digits (t-SNE 2D)")
plt.show()
```
